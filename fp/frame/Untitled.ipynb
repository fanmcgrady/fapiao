```python
lr = 0.25
p = torch.tensor([10., -10.], requires_grad=True)
x = torch.tensor([0.1, 0.2])
y = torch.tensor([0., 1.])
for i in range(100):
    y_ = p[1] * x + p[0]
    dy = y_ - y
    dy2 = dy**2
    loss = torch.mean(dy2)
    loss.backward()
    
    with torch.no_grad():
        p -= lr * p.grad
        print('{:8.4f} \t{} \t{} \t{}'.format(loss.item(), dy, p, p.grad))
        p.grad.zero_()
```
